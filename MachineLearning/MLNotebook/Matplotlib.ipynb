# 1. Basic

```python
import matplotlib.pyplot as plt
import numpy as np

# x is 50 points from -1 to 1
x = np.linspace(-1, 1, 50)
y = 2*x + 1 
y2 = x**2
plt.plot(x, y2)
plt.show()
```

## 1.1 Figure
    Different pics can stay in different figures.

```python
xn = np.linspace(-3, 3, 50)

# figure(num=3, figsize=(8,5))
plt.figure()
plt.plot(xn, y)

plt.figure()
plt.plot(xn, y)
plt.plot(xn, y2, color='red', linewidth=1.0, linestyle='--')
```

## 1.2 Aixs

```python
plt.plot(xn, y)
plt.xlim(-1,2)
plt.ylim(-2,3)
plt.xlabel('I am x')
plt.ylabel('I am y')

new_ticks = np.linspace(-1, 2, 5)
print(new_ticks)
plt.xticks(new_ticks)
plt.yticks([-2, -1.8, -1, 1.22, 3], [r"$really\ bad$", r"$bad$", r"$normal$", r"$good$", r"$really\ good$"])
```
